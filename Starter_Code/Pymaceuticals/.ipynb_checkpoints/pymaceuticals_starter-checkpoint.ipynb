```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_df = pd.merge(study_results, mouse_metadata, how='left', on='Mouse ID', copy=True)

# Display the data table for preview
combined_df.head()
```

```python
# Checking the number of mice.
unique_mice_series = pd.unique(combined_df['Mouse ID'])
len(unique_mice_series)
```

```python
type(unique_mice_series)
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
unique_mice_series
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
#Listing of Duplicate Data
## List title
print("List of Duplicate Data")
## initialize index for elements in list
count = 1
## initializes list of duplicate data list. Necessary for duplicate data df.
duplicate_data_list = []
## initialize loop from 0 to lenth of unique mice series to loop over all unique data
for mouse_index in range(len(unique_mice_series)):
    ### slice data for each unique mouse separately
    mouse_slice_df = combined_df.loc[combined_df['Mouse ID'] == unique_mice_series[mouse_index], :]
    #### if mouse_slice_df has duplicate data
    if len(mouse_slice_df) >1: 
        #### appends duplicate data to duplicate_data_list
        duplicate_data_list.append(mouse_slice_df)
        #### prints output of duplicate data
        ##### Creates output display of slice of mouse ID from unique_mice_series
        array_output= np.array(unique_mice_series[mouse_index], dtype='object')
        ##### prints index for elements in list of duplicate data
        print(f"{count}. \n")
        ##### increments index
        count+=1
        ##### prints output display of slice of mouse ID from unique_mice_series
        print(f"array([{array_output}], dtype={array_output.dtype})\n")
        ##### prints DataFrame of Duplicate Data
        print(mouse_slice_df)
        
        ##### prints new line character to maintain spacing for readibility
        print("\n")
    else:
        #### Passes to next mouse_index because there is nothing to be done.
        pass
    
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_data_df =pd.concat(duplicate_data_list, axis=0)
duplicate_data_df.head(10)
```

```python
# Display unique mice IDs 1
## turn series to list
mice_list = unique_mice_series.tolist()
print(len(mice_list))
## sort values by MouseID and Timepoint. 
sorted_combined_df = combined_df.sort_values(by=['Mouse ID', 'Timepoint'])
## reset the index to reference the elements of the DataFrame by order. Drop old indexes as a column
sorted_combined_df =sorted_combined_df.reset_index(drop=True)
# Initializes unique_mouse_df_list to hold individual unique_mouse data entires (a row)
unique_mouse_df_list = []
```

### Citation

1. Using Series.to_frame(): Series conversion to Dataframe
    - Url: https://pandas.pydata.org/docs/reference/api/pandas.Series.to_frame.html#pandas.Series.to_frame  
2. Transposing data from long format to wide format
    - pandas.DataFrame.T property (stores the transpose within the Dataframe instance)
    - Url: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.T.html

```python
for mouse in mice_list:
    mouse_slice_df = sorted_combined_df.loc[sorted_combined_df['Mouse ID'] == mouse, :]
    #print(unique_mouse_slice_df)
    mouse_slice_df = mouse_slice_df.reset_index(drop=True)
    unique_mouse_slice_series= mouse_slice_df.iloc[0, :]
    
    unique_mouse_slice_df = unique_mouse_slice_series.to_frame().T
    print(unique_mouse_slice_df)
    unique_mouse_df_list.append(unique_mouse_slice_df)
```

```python

# Create a clean DataFrame by dropping the duplicate mouse by its ID.
unique_combined_df = pd.concat(unique_mouse_df_list, axis=0, ignore_index=True)
unique_combined_df
```

```python
# Checking the number of mice in the clean DataFrame.
unique_combined_df.size // 8
```
